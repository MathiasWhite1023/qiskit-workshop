```python
def testar_dado_inteligente(num_jogadas=1000):
    """
    Teste completo do seu dado qu√¢ntico
    """
    print(f"üé≤ TESTE ESTAT√çSTICO DO DADO INTELIGENTE ({num_jogadas} jogadas)")
    print("=" * 60)
    
    try:
        # Coletar dados
        resultados = []
        print("üîÑ Coletando dados...")
        
        for i in range(num_jogadas):
            if (i + 1) % 200 == 0:
                print(f"   Progresso: {i + 1}/{num_jogadas}")
            
            resultado = dado_quantico_inteligente()
            resultados.append(resultado)
            
            # Verifica√ß√£o de seguran√ßa
            if resultado < 1 or resultado > 6:
                print(f"‚ùå ERRO: Dado retornou {resultado} (fora do range 1-6)!")
                return
        
        # An√°lise estat√≠stica
        contador = Counter(resultados)
        
        print(f"\nüìä RESULTADOS DETALHADOS:")
        print(f"{'Face':<6} {'Count':<8} {'Freq%':<8} {'Esperado%':<10} {'Diferen√ßa':<10} {'Status':<8}")
        print("-" * 60)
        
        esperado = num_jogadas / 6  # 16.67% cada
        
        for face in range(1, 7):
            count = contador.get(face, 0)
            freq_percent = (count / num_jogadas) * 100
            esperado_percent = 100 / 6
            diferenca = freq_percent - esperado_percent
            
            # Status visual
            if abs(diferenca) <= 2:
                status = "‚úÖ √ìtimo"
            elif abs(diferenca) <= 4:
                status = "üëç Bom"
            else:
                status = "‚ö†Ô∏è Revisar"
            
            print(f"{face:<6} {count:<8} {freq_percent:<7.1f}% {esperado_percent:<9.1f}% {diferenca:<+9.1f}% {status:<8}")
        
        # Visualiza√ß√£o
        faces = list(range(1, 7))
        frequencias = [contador.get(i, 0) for i in faces]
        
        plt.figure(figsize=(14, 6))
        
        # Gr√°fico de barras
        plt.subplot(1, 2, 1)
        colors = ['green' if abs(f - esperado) <= esperado*0.15 else 
                 'orange' if abs(f - esperado) <= esperado*0.3 else 'red' 
                 for f in frequencias]
        
        bars = plt.bar(faces, frequencias, color=colors, alpha=0.7, edgecolor='black')
        plt.axhline(y=esperado, color='red', linestyle='--', linewidth=2, 
                   label=f'Esperado ({esperado:.1f})')
        
        # Adicionar n√∫meros nas barras
        for bar, freq in zip(bars, frequencias):
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5, 
                    str(freq), ha='center', va='bottom', fontweight='bold')
        
        plt.xlabel('Face do Dado')
        plt.ylabel('Frequ√™ncia')
        plt.title(f'üé≤ Distribui√ß√£o do Dado Qu√¢ntico Inteligente\n({num_jogadas} jogadas)')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.ylim(0, max(frequencias) * 1.1)
        
        # Gr√°fico de desvios
        plt.subplot(1, 2, 2)
        desvios = [f - esperado for f in frequencias]
        colors_desvio = ['red' if d < -20 else 'green' if abs(d) <= 20 else 'orange' for d in desvios]
        
        bars2 = plt.bar(faces, desvios, color=colors_desvio, alpha=0.7)
        plt.axhline(y=0, color='black', linestyle='-', linewidth=2)
        plt.axhline(y=20, color='orange', linestyle='--', alpha=0.7, label='¬±20 (bom)')
        plt.axhline(y=-20, color='orange', linestyle='--', alpha=0.7)
        
        plt.xlabel('Face do Dado')
        plt.ylabel('Desvio da Frequ√™ncia Esperada')
        plt.title('üìà An√°lise de Desvios')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        # Avalia√ß√£o final
        max_desvio = max(abs(d) for d in desvios)
        
        print(f"\nüéØ AVALIA√á√ÉO FINAL:")
        print(f"   Maior desvio: ¬±{max_desvio:.1f} ocorr√™ncias")
        
        if max_desvio <= esperado * 0.15:
            print(f"   üèÜ EXCELENTE! Distribui√ß√£o quase perfeita!")
            print(f"   üéâ Seu dado qu√¢ntico est√° funcionando perfeitamente!")
        elif max_desvio <= esperado * 0.3:
            print(f"   ‚úÖ MUITO BOM! Dentro das flutua√ß√µes estat√≠sticas normais.")
            print(f"   üëç Seu dado est√° bem calibrado!")
        else:
            print(f"   ‚ö†Ô∏è BOM, mas pode melhorar. Verifique sua implementa√ß√£o.")
            print(f"   üí° Dica: Certifique-se de que est√° rejeitando 0 e 7 corretamente.")
        
        return resultados
        
    except Exception as e:
        print(f"‚ùå Erro durante o teste: {e}")
        print(f"üí° Certifique-se de implementar dado_quantico_inteligente() primeiro!")
        return None

# Execute o teste - descomente ap√≥s implementar sua fun√ß√£o
# resultados_dado = testar_dado_inteligente(1200)

print("üí° Implemente dado_quantico_inteligente() e descomente a linha acima para testar!")
```
