## âš¡ **EXERCÃCIO GUIADO 1: Porta CNOT - O XOR QuÃ¢ntico**

### ğŸ¯ **Objetivo**: Entender e usar a porta mais importante da computaÃ§Ã£o quÃ¢ntica!

**ğŸ¤” Pergunta**: O que Ã© uma porta controlada?

**ğŸ’¡ Resposta**: Ã‰ uma porta que age no qubit "alvo" apenas se o qubit "controle" estiver em |1âŸ©!

### ğŸ“š **CNOT (Controlled-NOT):**
- **Dois qubits**: Controle + Alvo
- **OperaÃ§Ã£o**: Se controle=1, aplica NOT no alvo
- **Tabela verdade**: |00âŸ©â†’|00âŸ©, |01âŸ©â†’|01âŸ©, |10âŸ©â†’|11âŸ©, |11âŸ©â†’|10âŸ©

```python
def demonstrar_cnot():
    """
    DEMONSTRAÃ‡ÃƒO: Como funciona a porta CNOT
    """
    print("âš¡ DEMONSTRAÃ‡ÃƒO: Porta CNOT (Controlled-NOT)")
    print("=" * 50)
    
    # Vamos testar todas as 4 combinaÃ§Ãµes possÃ­veis de entrada
    entradas = [
        (0, 0, "00: Nada muda"),
        (0, 1, "01: Nada muda (controle=0)"),
        (1, 0, "10: Alvo vira 1 (controle=1)"),
        (1, 1, "11: Alvo vira 0 (controle=1)")
    ]
    
    print(f"{'Entrada':<10} {'SaÃ­da':<10} {'ExplicaÃ§Ã£o':<25} {'VerificaÃ§Ã£o':<15}")
    print("-" * 65)
    
    for controle, alvo, descricao in entradas:
        print(f"\nğŸ”§ Testando entrada: |{controle}{alvo}âŸ© - {descricao}")
        
        # Criar circuito
        qc = QuantumCircuit(2, 2)  # 2 qubits, 2 bits clÃ¡ssicos
        
        # Inicializar qubits no estado desejado
        if controle == 1:
            qc.x(0)  # Qubit 0 = controle
            print(f"   Inicializado controle (q0) = |1âŸ©")
        
        if alvo == 1:
            qc.x(1)  # Qubit 1 = alvo
            print(f"   Inicializado alvo (q1) = |1âŸ©")
        
        # Aplicar CNOT: controle=0, alvo=1
        qc.cx(0, 1)  # cx = CNOT
        print(f"   âš¡ CNOT aplicada: controle(q0) â†’ alvo(q1)")
        
        # Medir
        qc.measure(0, 0)  # Qubit 0 â†’ bit 0
        qc.measure(1, 1)  # Qubit 1 â†’ bit 1
        
        # Executar
        simulador = AerSimulator()
        job = execute(qc, simulador, shots=1000)
        resultado = job.result()
        counts = resultado.get_counts(qc)
        
        # Resultado mais frequente (deveria ser 100%)
        resultado_str = max(counts, key=counts.get)
        
        print(f"   ğŸ¯ Resultado: |{resultado_str}âŸ© (frequÃªncia: {counts[resultado_str]}/1000)")
        
        # VerificaÃ§Ã£o lÃ³gica
        controle_final = int(resultado_str[0])
        alvo_final = int(resultado_str[1])
        
        # O controle nunca muda
        controle_ok = (controle_final == controle)
        
        # O alvo muda apenas se controle=1
        if controle == 1:
            alvo_esperado = 1 - alvo  # NOT do alvo original
        else:
            alvo_esperado = alvo  # Alvo nÃ£o muda
        
        alvo_ok = (alvo_final == alvo_esperado)
        
        status = "âœ…" if (controle_ok and alvo_ok) else "âŒ"
        
        print(f"   {status} Controle: {controle}â†’{controle_final}, Alvo: {alvo}â†’{alvo_final}")
        
        # Tabela resumo
        entrada_str = f"|{controle}{alvo}âŸ©"
        saida_str = f"|{controle_final}{alvo_final}âŸ©"
        print(f"{entrada_str:<10} {saida_str:<10} {descricao:<25} {status:<15}")
    
    print(f"\nğŸ¯ RESUMO DA PORTA CNOT:")
    print(f"   â€¢ Qubit controle NUNCA muda")
    print(f"   â€¢ Qubit alvo muda APENAS se controle=1")
    print(f"   â€¢ Ã‰ reversÃ­vel: CNOT(CNOT(x)) = x")
    print(f"   â€¢ Base de muitos algoritmos quÃ¢nticos!")

# Execute a demonstraÃ§Ã£o
demonstrar_cnot()
```

## ğŸ”„ **EXERCÃCIO GUIADO 2: Construindo XOR QuÃ¢ntico**

### ğŸ¯ **Objetivo**: Usar CNOT para implementar a operaÃ§Ã£o XOR

**ğŸ¤” Pergunta**: Como fazer A XOR B usando portas quÃ¢nticas?

**ğŸ’¡ Resposta**: CNOT jÃ¡ Ã© um XOR! Se inicializarmos o alvo como |0âŸ©, o resultado final serÃ¡ A XOR B!

```python
def xor_quantico(a, b):
    """
    IMPLEMENTAÃ‡ÃƒO GUIADA: XOR quÃ¢ntico usando CNOT
    
    Args:
        a, b (int): bits de entrada (0 ou 1)
    
    Returns:
        int: resultado do XOR (a âŠ• b)
    """
    print(f"ğŸ”„ Calculando {a} XOR {b} quanticamente...")
    
    # Passo 1: Criar circuito com 3 qubits
    # q0 = entrada A, q1 = entrada B, q2 = resultado
    qc = QuantumCircuit(3, 1)  # 3 qubits, 1 bit para resultado
    print(f"   âœ… Circuito criado: 3 qubits (A, B, resultado)")
    
    # Passo 2: Inicializar entradas A e B
    if a == 1:
        qc.x(0)  # Qubit 0 = A
        print(f"   âœ… Entrada A inicializada como |1âŸ©")
    
    if b == 1:
        qc.x(1)  # Qubit 1 = B
        print(f"   âœ… Entrada B inicializada como |1âŸ©")
    
    # Passo 3: Aplicar CNOT para implementar XOR
    # XOR = A CNOT resultado + B CNOT resultado
    qc.cx(0, 2)  # A XOR resultado â†’ resultado
    qc.cx(1, 2)  # B XOR resultado â†’ resultado
    print(f"   âš¡ CNOT aplicadas: AâŠ•resultado, BâŠ•resultado")
    print(f"   ğŸ§® MatemÃ¡tica: (0âŠ•A)âŠ•B = AâŠ•B")
    
    # Passo 4: Medir apenas o resultado
    qc.measure(2, 0)  # Qubit resultado â†’ bit clÃ¡ssico
    print(f"   ğŸ“ Medindo apenas o qubit resultado")
    
    # Passo 5: Executar
    simulador = AerSimulator()
    job = execute(qc, simulador, shots=1)
    resultado_job = job.result()
    counts = resultado_job.get_counts(qc)
    
    resultado = int(list(counts.keys())[0])
    
    print(f"   ğŸ¯ Resultado quÃ¢ntico: {resultado}")
    
    # VerificaÃ§Ã£o clÃ¡ssica
    xor_classico = a ^ b  # XOR clÃ¡ssico em Python
    print(f"   âœ… VerificaÃ§Ã£o clÃ¡ssica: {a} âŠ• {b} = {xor_classico}")
    
    if resultado == xor_classico:
        print(f"   ğŸ‰ SUCESSO! XOR quÃ¢ntico funcionou perfeitamente!")
    else:
        print(f"   âŒ Erro: esperado {xor_classico}, obtido {resultado}")
    
    return resultado

# Teste completo do XOR quÃ¢ntico
print("ğŸ”„ TESTE COMPLETO: XOR QuÃ¢ntico vs. ClÃ¡ssico")
print("=" * 50)

print(f"\n{'A':<3} {'B':<3} {'ClÃ¡ssico':<10} {'QuÃ¢ntico':<10} {'Status':<8}")
print("-" * 35)

for a in [0, 1]:
    for b in [0, 1]:
        print(f"\nğŸ§ª Teste: A={a}, B={b}")
        resultado_quantico = xor_quantico(a, b)
        resultado_classico = a ^ b
        status = "âœ…" if resultado_quantico == resultado_classico else "âŒ"
        
        print(f"{a:<3} {b:<3} {resultado_classico:<10} {resultado_quantico:<10} {status:<8}")

print(f"\nğŸ¯ XOR QuÃ¢ntico implementado com sucesso usando CNOT!")
```

## ğŸ—ï¸ **EXERCÃCIO GUIADO 3: Porta Toffoli - O AND ReversÃ­vel**

### ğŸ¯ **Objetivo**: Implementar AND quÃ¢ntico usando a famosa porta Toffoli

**ğŸ¤” Pergunta**: Por que AND Ã© problemÃ¡tico na computaÃ§Ã£o quÃ¢ntica?

**ğŸ’¡ Resposta**: AND nÃ£o Ã© reversÃ­vel! De |11âŸ©â†’|1âŸ© vocÃª nÃ£o consegue voltar. A porta Toffoli resolve isso usando um qubit auxiliar!

### ğŸ“š **Porta Toffoli (CCX):**
- **3 qubits**: Controle1, Controle2, Alvo
- **OperaÃ§Ã£o**: Alvo muda APENAS se ambos controles = 1
- **ReversÃ­vel**: Aplicar duas vezes retorna ao estado original

```python
def and_quantico_reversivel(a, b):
    """
    IMPLEMENTAÃ‡ÃƒO GUIADA: AND quÃ¢ntico usando porta Toffoli
    
    Args:
        a, b (int): bits de entrada (0 ou 1)
    
    Returns:
        int: resultado do AND (a & b)
    """
    print(f"ğŸ—ï¸ Calculando {a} AND {b} com porta Toffoli...")
    
    # Passo 1: Criar circuito com 3 qubits
    # q0 = entrada A, q1 = entrada B, q2 = auxiliar/resultado
    qc = QuantumCircuit(3, 1)
    print(f"   âœ… Circuito criado: 3 qubits (A, B, auxiliar)")
    
    # Passo 2: Inicializar entradas
    if a == 1:
        qc.x(0)
        print(f"   âœ… Entrada A = |1âŸ©")
    
    if b == 1:
        qc.x(1)
        print(f"   âœ… Entrada B = |1âŸ©")
    
    # Qubit auxiliar jÃ¡ inicia como |0âŸ©
    print(f"   âœ… Auxiliar = |0âŸ© (estado inicial)")
    
    # Passo 3: Aplicar porta Toffoli (CCX)
    qc.ccx(0, 1, 2)  # ccx = Toffoli = Controlled-Controlled-X
    print(f"   âš¡ Toffoli aplicada: se A=1 E B=1, entÃ£o auxiliar vira 1")
    print(f"   ğŸ§® LÃ³gica: auxiliar = auxiliar âŠ• (A âˆ§ B) = 0 âŠ• (A âˆ§ B) = A âˆ§ B")
    
    # Passo 4: Medir o qubit auxiliar (contÃ©m o resultado)
    qc.measure(2, 0)
    print(f"   ğŸ“ Medindo qubit auxiliar (resultado AND)")
    
    # Passo 5: Executar
    simulador = AerSimulator()
    job = execute(qc, simulador, shots=1)
    resultado_job = job.result()
    counts = resultado_job.get_counts(qc)
    
    resultado = int(list(counts.keys())[0])
    
    print(f"   ğŸ¯ Resultado quÃ¢ntico: {resultado}")
    
    # VerificaÃ§Ã£o
    and_classico = a & b
    print(f"   âœ… VerificaÃ§Ã£o clÃ¡ssica: {a} âˆ§ {b} = {and_classico}")
    
    if resultado == and_classico:
        print(f"   ğŸ‰ SUCESSO! AND reversÃ­vel funcionou!")
    else:
        print(f"   âŒ Erro inesperado")
    
    return resultado

def demonstrar_reversibilidade():
    """
    DEMONSTRAÃ‡ÃƒO: Por que a porta Toffoli Ã© reversÃ­vel
    """
    print(f"\nğŸ”„ DEMONSTRAÃ‡ÃƒO: Reversibilidade da Porta Toffoli")
    print(f"=" * 55)
    
    # Exemplo: comeÃ§ar com |110âŸ© (A=1, B=1, auxiliar=0)
    print(f"ğŸ§ª Teste de reversibilidade: |110âŸ© â†’ Toffoli â†’ Toffoli â†’ |110âŸ©")
    
    qc = QuantumCircuit(3, 3)
    
    # Estado inicial: |110âŸ©
    qc.x(0)  # A = 1
    qc.x(1)  # B = 1
    # auxiliar = 0 (padrÃ£o)
    
    # Primeira Toffoli
    qc.ccx(0, 1, 2)
    print(f"   ApÃ³s 1Âª Toffoli: |110âŸ© â†’ |111âŸ© (auxiliar virou 1)")
    
    # Segunda Toffoli (desfaz a primeira!)
    qc.ccx(0, 1, 2)
    print(f"   ApÃ³s 2Âª Toffoli: |111âŸ© â†’ |110âŸ© (voltou ao original!)")
    
    # Medir tudo
    qc.measure_all()
    
    simulador = AerSimulator()
    job = execute(qc, simulador, shots=1000)
    resultado = job.result()
    counts = resultado.get_counts(qc)
    
    print(f"   ğŸ¯ Resultado final: {counts}")
    print(f"   âœ… Se '110' aparece ~1000 vezes, a reversibilidade estÃ¡ confirmada!")

# Teste completo do AND quÃ¢ntico
print("ğŸ—ï¸ TESTE COMPLETO: AND QuÃ¢ntico ReversÃ­vel")
print("=" * 50)

print(f"\n{'A':<3} {'B':<3} {'ClÃ¡ssico':<10} {'QuÃ¢ntico':<10} {'Status':<8}")
print("-" * 35)

for a in [0, 1]:
    for b in [0, 1]:
        print(f"\nğŸ§ª Teste: A={a}, B={b}")
        resultado_quantico = and_quantico_reversivel(a, b)
        resultado_classico = a & b
        status = "âœ…" if resultado_quantico == resultado_classico else "âŒ"
        
        print(f"{a:<3} {b:<3} {resultado_classico:<10} {resultado_quantico:<10} {status:<8}")

# Demonstrar reversibilidade
demonstrar_reversibilidade()

print(f"\nğŸ¯ Porta Toffoli: A base do AND quÃ¢ntico reversÃ­vel!")
```

## ğŸšï¸ **EXERCÃCIO GUIADO 4: RotaÃ§Ãµes QuÃ¢nticas - Controlando Probabilidades**

### ğŸ¯ **Objetivo**: Aprender a controlar probabilidades precisamente usando rotaÃ§Ãµes

**ğŸ¤” Pergunta**: Como criar uma moeda com probabilidade especÃ­fica (ex: 30% cara)?

**ğŸ’¡ Resposta**: Usando rotaÃ§Ãµes na esfera de Bloch! As portas RY permitem controle preciso das probabilidades.

### ğŸ“š **RotaÃ§Ãµes QuÃ¢nticas:**
- **RX(Î¸)**: RotaÃ§Ã£o no eixo X
- **RY(Î¸)**: RotaÃ§Ã£o no eixo Y (mais usada para probabilidades)
- **RZ(Î¸)**: RotaÃ§Ã£o no eixo Z

### ğŸ§® **MatemÃ¡tica:**
Para probabilidade p de medir |1âŸ©: **Î¸ = 2 Ã— arcsin(âˆšp)**

```python
def moeda_personalizada(probabilidade_cara):
    """
    IMPLEMENTAÃ‡ÃƒO GUIADA: Moeda com probabilidade customizada
    
    Args:
        probabilidade_cara (float): probabilidade desejada (0.0 a 1.0)
    
    Returns:
        int: 0 (coroa) ou 1 (cara)
    """
    print(f"ğŸšï¸ Criando moeda com {probabilidade_cara*100:.1f}% de chance de cara...")
    
    # Passo 1: Calcular o Ã¢ngulo de rotaÃ§Ã£o
    # Para probabilidade p: Î¸ = 2 * arcsin(âˆšp)
    theta = 2 * np.arcsin(np.sqrt(probabilidade_cara))
    print(f"   ğŸ§® Ã‚ngulo calculado: Î¸ = {theta:.4f} radianos ({np.degrees(theta):.1f}Â°)")
    
    # Passo 2: Criar circuito
    qc = QuantumCircuit(1, 1)
    print(f"   âœ… Circuito criado")
    
    # Passo 3: Aplicar rotaÃ§Ã£o RY
    qc.ry(theta, 0)
    print(f"   âš¡ RotaÃ§Ã£o RY({theta:.4f}) aplicada")
    print(f"   ğŸŒŸ Qubit agora tem {probabilidade_cara*100:.1f}% chance de ser |1âŸ©")
    
    # Passo 4: Medir
    qc.measure(0, 0)
    
    # Passo 5: Executar
    simulador = AerSimulator()
    job = execute(qc, simulador, shots=1)
    resultado = job.result()
    counts = resultado.get_counts(qc)
    
    resultado_bit = int(list(counts.keys())[0])
    nome_resultado = "Cara" if resultado_bit == 1 else "Coroa"
    
    print(f"   ğŸ¯ Resultado: {resultado_bit} ({nome_resultado})")
    
    return resultado_bit

def testar_moeda_personalizada(probabilidade, num_testes=1000):
    """
    Teste estatÃ­stico da moeda personalizada
    """
    print(f"\nğŸ§ª TESTE ESTATÃSTICO: Moeda {probabilidade*100:.1f}% cara ({num_testes} jogadas)")
    print("=" * 60)
    
    # VersÃ£o otimizada para mÃºltiplos testes
    theta = 2 * np.arcsin(np.sqrt(probabilidade))
    
    qc = QuantumCircuit(1, 1)
    qc.ry(theta, 0)
    qc.measure(0, 0)
    
    print(f"ğŸ° Executando {num_testes} jogadas simultaneamente...")
    
    simulador = AerSimulator()
    job = execute(qc, simulador, shots=num_testes)
    resultado = job.result()
    counts = resultado.get_counts(qc)
    
    # AnÃ¡lise dos resultados
    coroas = counts.get('0', 0)
    caras = counts.get('1', 0)
    
    freq_caras_observada = caras / num_testes
    erro = abs(freq_caras_observada - probabilidade)
    
    print(f"\nğŸ“Š RESULTADOS:")
    print(f"   ğŸª™ Coroas (0): {coroas} ({coroas/num_testes*100:.1f}%)")
    print(f"   ğŸª™ Caras (1):  {caras} ({caras/num_testes*100:.1f}%)")
    print(f"   ğŸ“ˆ Total: {coroas + caras}")
    
    print(f"\nğŸ¯ ANÃLISE:")
    print(f"   Probabilidade esperada: {probabilidade*100:.1f}%")
    print(f"   Probabilidade observada: {freq_caras_observada*100:.1f}%")
    print(f"   Erro absoluto: {erro*100:.2f}%")
    
    # AvaliaÃ§Ã£o
    if erro <= 0.02:  # Â±2%
        print(f"   âœ… EXCELENTE! PrecisÃ£o muito alta")
    elif erro <= 0.05:  # Â±5%
        print(f"   ğŸ‘ MUITO BOM! Dentro da tolerÃ¢ncia")
    else:
        print(f"   âš ï¸ Erro maior que 5% - pode ser flutuaÃ§Ã£o estatÃ­stica")
    
    return {
        'coroas': coroas,
        'caras': caras,
        'freq_observada': freq_caras_observada,
        'erro': erro
    }

# DemonstraÃ§Ã£o com diferentes probabilidades
print("ğŸšï¸ DEMONSTRAÃ‡ÃƒO: Controle Preciso de Probabilidades")
print("=" * 60)

probabilidades_teste = [0.2, 0.5, 0.8]  # 20%, 50%, 80%

print(f"\nğŸ§ª Teste individual (uma jogada cada):")
for prob in probabilidades_teste:
    print(f"\nğŸ“Š Testando probabilidade {prob*100:.0f}%:")
    resultado = moeda_personalizada(prob)

# Teste estatÃ­stico completo
print(f"\n\nğŸ”¬ ANÃLISE ESTATÃSTICA COMPLETA:")
resultados = []

for prob in probabilidades_teste:
    resultado_teste = testar_moeda_personalizada(prob, 1000)
    resultados.append((prob, resultado_teste))

# VisualizaÃ§Ã£o comparativa
print(f"\nğŸ“ˆ Criando visualizaÃ§Ã£o comparativa...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# GrÃ¡fico 1: Probabilidades esperadas vs observadas
probs_esperadas = [r[0] for r in resultados]
probs_observadas = [r[1]['freq_observada'] for r in resultados]

ax1.scatter(probs_esperadas, probs_observadas, s=100, alpha=0.7, color='blue')
ax1.plot([0, 1], [0, 1], 'r--', alpha=0.7, label='Perfeito (x=y)')

for i, (esp, obs) in enumerate(zip(probs_esperadas, probs_observadas)):
    ax1.annotate(f'{esp*100:.0f}%', (esp, obs), xytext=(5, 5), 
                textcoords='offset points', fontsize=10)

ax1.set_xlabel('Probabilidade Esperada')
ax1.set_ylabel('Probabilidade Observada')
ax1.set_title('ğŸ¯ PrecisÃ£o das RotaÃ§Ãµes QuÃ¢nticas')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)

# GrÃ¡fico 2: Erro por probabilidade
erros = [r[1]['erro'] for r in resultados]
ax2.bar(range(len(probabilidades_teste)), [e*100 for e in erros], 
        alpha=0.7, color=['green' if e <= 0.02 else 'orange' for e in erros])

ax2.set_xlabel('Probabilidade Testada')
ax2.set_ylabel('Erro Absoluto (%)')
ax2.set_title('ğŸ“Š Erro por ConfiguraÃ§Ã£o')
ax2.set_xticks(range(len(probabilidades_teste)))
ax2.set_xticklabels([f'{p*100:.0f}%' for p in probabilidades_teste])
ax2.axhline(y=2, color='red', linestyle='--', alpha=0.7, label='Meta (Â±2%)')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nğŸ‰ RotaÃ§Ãµes quÃ¢nticas permitem controle preciso de probabilidades!")
print(f"ğŸŒŸ Isso Ã© fundamental para algoritmos quÃ¢nticos avanÃ§ados!")
```

## ğŸ§¬ **EXERCÃCIO GUIADO 5: EntrelaÃ§amento QuÃ¢ntico BÃ¡sico**

### ğŸ¯ **Objetivo**: Criar e explorar o fenÃ´meno mais "mÃ¡gico" da mecÃ¢nica quÃ¢ntica!

**ğŸ¤” Pergunta**: O que Ã© entrelaÃ§amento quÃ¢ntico?

**ğŸ’¡ Resposta**: Ã‰ quando dois qubits ficam "conectados" de forma que medir um afeta instantaneamente o outro, nÃ£o importa a distÃ¢ncia!

### ğŸ§¬ **Estados de Bell:**
Estados entrelaÃ§ados fundamentais:
- **|Î¦âºâŸ© = (|00âŸ© + |11âŸ©)/âˆš2**: "Mesma paridade"
- **|Î¨âºâŸ© = (|01âŸ© + |10âŸ©)/âˆš2**: "Paridade oposta"

```python
def criar_estado_bell_phi_plus():
    """
    IMPLEMENTAÃ‡ÃƒO GUIADA: Estado de Bell |Î¦âºâŸ© = (|00âŸ© + |11âŸ©)/âˆš2
    
    Este Ã© o estado entrelaÃ§ado mais famoso!
    Propriedade: Se um qubit Ã© 0, o outro SEMPRE Ã© 0.
                 Se um qubit Ã© 1, o outro SEMPRE Ã© 1.
    """
    print("ğŸ§¬ CRIANDO ESTADO DE BELL |Î¦âºâŸ©")
    print("=" * 40)
    
    # Passo 1: Criar circuito com 2 qubits
    qc = QuantumCircuit(2, 2)
    print("   âœ… Circuito criado: 2 qubits")
    
    # Passo 2: Estado inicial |00âŸ©
    print("   ğŸ“ Estado inicial: |00âŸ©")
    
    # Passo 3: Aplicar Hadamard no primeiro qubit
    qc.h(0)
    print("   âš¡ Hadamard no qubit 0: |00âŸ© â†’ (|00âŸ© + |10âŸ©)/âˆš2")
    print("      ğŸŒŸ Agora qubit 0 estÃ¡ em superposiÃ§Ã£o!")
    
    # Passo 4: Aplicar CNOT
    qc.cx(0, 1)
    print("   âš¡ CNOT(0â†’1): (|00âŸ© + |10âŸ©)/âˆš2 â†’ (|00âŸ© + |11âŸ©)/âˆš2")
    print("   ğŸ§¬ ENTRELAÃ‡AMENTO CRIADO!")
    print("      ğŸ’« Os qubits agora estÃ£o 'conectados' quanticamente!")
    
    # Passo 5: Medir
    qc.measure_all()
    
    # Passo 6: Executar mÃºltiplas vezes para ver o padrÃ£o
    print("\nğŸ° Executando 1000 mediÃ§Ãµes para observar entrelaÃ§amento...")
    
    simulador = AerSimulator()
    job = execute(qc, simulador, shots=1000)
    resultado = job.result()
    counts = resultado.get_counts(qc)
    
    print(f"\nğŸ“Š RESULTADOS:")
    for estado, freq in counts.items():
        percentage = freq / 1000 * 100
        print(f"   |{estado}âŸ©: {freq} vezes ({percentage:.1f}%)")
    
    # AnÃ¡lise do entrelaÃ§amento
    print(f"\nğŸ” ANÃLISE DO ENTRELAÃ‡AMENTO:")
    
    zeros_zeros = counts.get('00', 0)
    uns_uns = counts.get('11', 0)
    zeros_uns = counts.get('01', 0)
    uns_zeros = counts.get('10', 0)
    
    print(f"   Estados 'correlacionados' (00 + 11): {zeros_zeros + uns_uns} ({(zeros_zeros + uns_uns)/10:.1f}%)")
    print(f"   Estados 'descorrelacionados' (01 + 10): {zeros_uns + uns_zeros} ({(zeros_uns + uns_zeros)/10:.1f}%)")
    
    if zeros_uns + uns_zeros == 0:
        print(f"   âœ… ENTRELAÃ‡AMENTO PERFEITO! Apenas estados correlacionados aparecem!")
        print(f"   ğŸŒŸ Os qubits sempre tÃªm o mesmo valor!")
    elif zeros_uns + uns_zeros < 50:
        print(f"   ğŸ‘ ENTRELAÃ‡AMENTO FORTE! CorrelaÃ§Ã£o muito alta.")
    else:
        print(f"   âš ï¸ PossÃ­vel erro - deveria haver apenas 00 e 11")
    
    return qc, counts

def demonstrar_correlacao_quantica():
    """
    DEMONSTRAÃ‡ÃƒO: Como funciona a correlaÃ§Ã£o no entrelaÃ§amento
    """
    print(f"\n\nğŸ”¬ DEMONSTRAÃ‡ÃƒO: CorrelaÃ§Ã£o QuÃ¢ntica InstantÃ¢nea")
    print(f"=" * 50)
    
    print(f"ğŸ§ª Experimento: Medir qubits entrelaÃ§ados separadamente")
    print(f"ğŸ’« Se medirmos apenas UM qubit, o outro fica automaticamente determinado!")
    
    # Criar estado entrelaÃ§ado
    qc = QuantumCircuit(2, 2)
    qc.h(0)      # SuperposiÃ§Ã£o
    qc.cx(0, 1)  # EntrelaÃ§amento
    
    # Experimento 1: Medir apenas o primeiro qubit
    qc1 = qc.copy()
    qc1.measure(0, 0)  # Medir APENAS qubit 0
    
    print(f"\nğŸ¯ Teste 1: Medindo APENAS o qubit 0")
    
    simulador = AerSimulator()
    job1 = execute(qc1, simulador, shots=100)
    resultado1 = job1.result()
    counts1 = resultado1.get_counts(qc1)
    
    print(f"   Resultados do qubit 0: {counts1}")
    print(f"   ğŸ“Š Como esperado: ~50% para cada valor (qubit 0 aleatÃ³rio)")
    
    # Experimento 2: Medir ambos os qubits
    qc2 = qc.copy()
    qc2.measure_all()
    
    print(f"\nğŸ¯ Teste 2: Medindo AMBOS os qubits")
    
    job2 = execute(qc2, simulador, shots=100)
    resultado2 = job2.result()
    counts2 = resultado2.get_counts(qc2)
    
    print(f"   Resultados de ambos: {counts2}")
    print(f"   ğŸ§¬ MAGIA QUÃ‚NTICA: Apenas 00 e 11 aparecem!")
    print(f"   ğŸ’« Os qubits estÃ£o perfeitamente correlacionados!")
    
    # VisualizaÃ§Ã£o
    print(f"\nğŸ“ˆ Criando visualizaÃ§Ã£o da correlaÃ§Ã£o...")
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # GrÃ¡fico 1: MediÃ§Ã£o individual (qubit 0)
    if counts1:
        estados1 = list(counts1.keys())
        freq1 = list(counts1.values())
        
        ax1.bar(estados1, freq1, alpha=0.7, color='blue')
        ax1.set_title('ğŸ“Š MediÃ§Ã£o Individual\n(Apenas Qubit 0)')
        ax1.set_xlabel('Estado do Qubit 0')
        ax1.set_ylabel('FrequÃªncia')
        ax1.grid(True, alpha=0.3)
    
    # GrÃ¡fico 2: MediÃ§Ã£o conjunta (ambos qubits)
    if counts2:
        estados2 = list(counts2.keys())
        freq2 = list(counts2.values())
        
        colors = ['green' if s in ['00', '11'] else 'red' for s in estados2]
        
        ax2.bar(estados2, freq2, alpha=0.7, color=colors)
        ax2.set_title('ğŸ§¬ MediÃ§Ã£o EntrelaÃ§ada\n(Ambos Qubits)')
        ax2.set_xlabel('Estado dos Qubits')
        ax2.set_ylabel('FrequÃªncia')
        ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nğŸ¯ CONCLUSÃƒO SURPREENDENTE:")
    print(f"   â€¢ Individualmente: cada qubit parece aleatÃ³rio (50/50)")
    print(f"   â€¢ Juntos: correlaÃ§Ã£o perfeita (sempre iguais)!")
    print(f"   ğŸŒŸ Isso Ã© o que Einstein chamou de 'aÃ§Ã£o fantasmagÃ³rica Ã  distÃ¢ncia'")

# Executar demonstraÃ§Ã£o completa
circuito_bell, resultados_bell = criar_estado_bell_phi_plus()
demonstrar_correlacao_quantica()

print(f"\nğŸ§¬ VocÃª acabou de criar e observar entrelaÃ§amento quÃ¢ntico!")
print(f"ğŸ‰ Este Ã© um dos fenÃ´menos mais fascinantes da fÃ­sica moderna!")
```
